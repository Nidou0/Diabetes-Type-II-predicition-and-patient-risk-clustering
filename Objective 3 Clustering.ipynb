```python
df.drop(columns=['Unnamed: 0'], inplace= True)
```

```python
df.head()
```

```python
clustering_features = df.drop(columns=['Diabetes'])
```

```python
from sklearn.cluster import KMeans
inertia = []
for k in range (1,10):
    kmeans = KMeans(n_clusters= k, random_state= 42)
    kmeans.fit(clustering_features)
    inertia.append(kmeans.inertia_)

plt.plot(range(1,10), inertia, marker = 'o')
plt.title('Elbow method')
plt.xlabel('Cluster')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters= 4, random_state=42)
df['Cluster'] = kmeans.fit_predict(clustering_features)
```

```python
cluster_risk = df.groupby('Cluster')['Diabetes'].mean()
print(cluster_risk)
```

```python
cluster_to_risk = {
    0: 'Very Low Risk',
    1: 'High Risk',
    2:'Low Risk',
    3: 'Moderate Risk'
}

risk_to_cluster = {
    'Very Low Risk': 0,
    'Low Risk' : 2,
    'Moderate Risk': 3,
    'High Risk': 1
}

df['Risk_level'] = df['Cluster'].map(cluster_to_risk)
df['Risk_score'] = df['Risk_level'].map(risk_to_cluster)
```

```python
df.head()
```

```python
sns.countplot(x = 'Cluster', data = df)
```

```python
from sklearn.metrics import silhouette_score

# Calculate silhouette score to evaluate clustering quality
sil_score = silhouette_score(clustering_features, df['Cluster'])
print(f"Silhouette Score: {sil_score:.4f}")

# Display cluster risk means for context
print("Cluster risk means:")
print(cluster_risk)
```
